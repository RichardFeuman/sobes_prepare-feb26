```python
from typing import Callable
import time
```

```python
def deco(func: Callable):
    def wrapper():
        start = time.time()
        res = func()
        end = time.time()
        print(f"время выполнения {end - start}")
        return res
    return wrapper
```

```python



def neu_network():
    pred = 0.85 * 0.5
    return pred


neu_network = deco(neu_network)
```

```python
neu_network()
```

```python
def param_deco(func: Callable):
    def wrapper(*args, **kwargs):
        start = time.time()
        res = func(*args, **kwargs)
        end = time.time()
        print(f"время выполнения {end - start}")
        return res
    return wrapper
```

```python

@param_deco
def neu_network2(input, weight):
    pred = input * weight
    return pred
```

```python
neu_network2(0.85, 0.5)
```

```python
from functools import wraps
```

```python
def limit_calls(limit: int):

    def wrapper(func):
      @wraps(func)
      def inner(*args, **kwargs):
        nonlocal limit
        if limit == 0:
          raise Exception("превышено число вызовов")
        res = func(*args, **kwargs)
        limit-=1
        return res
      return inner

    return wrapper
```

```python

@limit_calls(3)
def neu_network2(input, weight):
    "simple neural network"
    pred = input * weight
    return pred
```

```python
neu_network2(0.5, 0.5)
```

```python
neu_network2(0.25, 0.25)
```

```python
neu_network2(0.35, 0.35)
```

Декоратор для асинхронных функций

```python
from typing import Coroutine
```

```python
def deco(crt: Coroutine):
  async def wrapper(*args, **kwargs):
    res = await crt(*args, **kwargs)
    return res
    return wrapper
```

```python
async def my_async_func():
  return 1
```

```python
@deco
async def my_async_func():
  return 1
```

```python
await my_async_func()
```

Декоратор для создания контекстного менеджера

```python
from contextlib import contextmanager

@contextmanager
def my_manager():
    print('star')
    yield
    print('wars')
```

```python
with my_manager() as man:
  print('hi')
```
